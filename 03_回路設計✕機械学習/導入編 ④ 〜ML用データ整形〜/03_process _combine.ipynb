```python
import pandas as pd
import numpy as np

# プリフィックス
prefix = './data/01_single/output/'

# リファレンス一覧
ref = ["C1","C2","C3","C4","C5","C6","C7","C8",
        "C9","C10","C11","C12","C13","L1","L2","L3"]

# ファイルパス
filein = []
fileout = prefix + "ADS_out.csv"
for name in ref:
    filein.append(prefix + name + "_out.csv")

# データフレーム加工
def csv_process(file1, name1):
    df = pd.read_csv(file1)
    df['Unnamed: 0'] = name1
    return df

# データフレームの縦結合
dfout = pd.DataFrame()
for i in range(len(filein)):
    df1 = csv_process(filein[i], ref[i])
    dfout = pd.concat([dfout,df1], axis=0, ignore_index=True)

# ファイル出力
dfout.to_csv(fileout)
```
