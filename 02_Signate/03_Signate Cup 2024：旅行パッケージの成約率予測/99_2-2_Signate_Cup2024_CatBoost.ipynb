```python
# ãƒ©ã‚¤ãƒ–ãƒ©ãƒªã®ã‚¤ãƒ³ãƒãƒ¼ãƒˆ
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# ãƒ‡ãƒ¼ã‚¿ãƒ•ãƒ¬ãƒ¼ãƒ èª­ã¿è¾¼ã¿
train_df = pd.read_csv("/content/train.csv")
test_df = pd.read_csv("/content/test.csv")

# ãƒ‡ãƒ¼ã‚¿çµåˆ(ç•ªå·ã‚’æŒ¯ã‚Šç›´ã™)
all_df = pd.concat([train_df,test_df],axis=0,ignore_index=True)
max_row = len(all_df)
```

```python
###############
# "Age"ã®å‡¦ç† #
###############
import unicodedata # å…¨è§’ã‚’åŠè§’ã«å¤‰æ›
from kanjize import kanji2number # æ¼¢å­—ã‚’æ•°å€¤ã«å¤‰æ›
import re

# æ¼¢å­—ã®åˆ¤å®šç”¨
p = re.compile('[ä¸€äºŒä¸‰å››äº”å…­ä¸ƒå…«ä¹å]+')

# nanã‚’-1æ­³ã§è£œå®Œ
all_df["Age"] = all_df["Age"].fillna("-1")

# Ageã®å‡¦ç†
for i in range(len(all_df)):
# for i in range(max_row):

  # æœ«å°¾æ–‡å­—ã®å–å¾—
  tail = all_df.loc[i,"Age"][-1:]

  # å…ˆé ­æ–‡å­—ãŒæ¼¢å­—ã®å ´åˆ
  if(p.search(all_df.loc[i,"Age"][0])!=None):
    tmp = all_df.loc[i,"Age"][:-1]
    all_df.loc[i,"Age"] = kanji2number(tmp)
    all_df.loc[i,"Age"] = int(all_df.loc[i,"Age"])

  # æœ«å°¾æ–‡å­—ãŒæ­³ã€æ‰ã€éš›ã®å ´åˆ
  elif((tail=="æ­³")or(tail=="æ‰")or(tail=="éš›")):
    all_df.loc[i,"Age"] = all_df.loc[i,"Age"][:-1]
    # å…¨è§’ã‚’åŠè§’ã«ç›´ã™
    all_df.loc[i,"Age"] = unicodedata.normalize("NFKC", all_df.loc[i,"Age"])
    all_df.loc[i,"Age"] = int(all_df.loc[i,"Age"])

  # ä»£ã®å ´åˆ
  elif(tail == "ä»£"):
    tmp = all_df.loc[i,"Age"][:-1]

    # ã€Œä»£ã€ã¯â—‹5æ­³ã«å¤‰æ›ã™ã‚‹
    tmp = int(tmp)
    # 10ä»£ã¯19æ­³ã«ã™ã‚‹
    if(-1<tmp & tmp<20):
      all_df.loc[i,"Age"] = 19
    elif(tmp>60):
      all_df.loc[i,"Age"] = 61
    else:
      # 10æ­³ã§åˆ†ã‘ã¦5è¶³ã™
      tmp = tmp//10 * 10
      all_df.loc[i,"Age"] = tmp + 5

# intå‹ã«å¤‰æ›
all_df["Age"] = all_df["Age"].astype(int)

print(all_df.groupby("Age").size())
```

```python
##########################
# "ProductPitched"ã®å‡¦ç† #
##########################
# ã‚®ãƒªã‚·ãƒ£æ–‡å­—ã‚’ã‚¢ãƒ«ãƒ•ã‚¡ãƒ™ãƒƒãƒˆã«å¤‰æ›
greek_string = "Î±Î²Î³Î´ÎµÎ¶Î·Î¸Î¹ÎºÎ»Î¼Î½Î¾Î¿Ï€ÏÏƒÏ„Ï…Ï†Ï‡ÏˆÏ‰Ğ²ê“¢Ñ•ê­°Õ¿á—Ä±ğŠ¡"
alphabet_string = "abgdezhqiklmnxoprstufcywbssdsdib"

def greek_to_alphabet(text):
    result = ""
    for char in text:
        if char in greek_string:
            result += alphabet_string[greek_string.index(char)]
        else:
            result += char  # ã‚®ãƒªã‚·ãƒ£æ–‡å­—ä»¥å¤–ã®æ–‡å­—ã¯ãã®ã¾ã¾
    return result

# Basic, Standard, Deluxe, Super Deluxe, kingã«åˆ†ã‘ã‚‹
prefix = 0
others = []
# ã™ã¹ã¦å°æ–‡å­—ã«å¤‰æ›ã—åˆ¤å®š
for i in range(len(all_df)):
  prefix = unicodedata.normalize("NFKC", greek_to_alphabet(all_df.loc[i,"ProductPitched"][:2].lower())).lower()

  if(prefix == "ba"):
    all_df.loc[i,"ProductPitched"] = "Basic"

  elif(prefix == "st"):
    all_df.loc[i,"ProductPitched"] = "Standard"

  elif(prefix == "de"):
    all_df.loc[i,"ProductPitched"] = "Deluxe"

  elif(prefix == "su"):
    all_df.loc[i,"ProductPitched"] = "SuperDeluxe"

  elif(prefix == "ki"):
    all_df.loc[i,"ProductPitched"] = "King"

  else:
    others.append(i)

all_df["ProductPitched"] = all_df["ProductPitched"].astype("object")

# åˆ†å¸ƒã®ç¢ºèª
print(all_df.groupby(["ProductPitched"]).size())
```
